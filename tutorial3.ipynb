ðŸ“ƒ **4**. Create a random numpy array of size 6x3 and transpose it

```python
arr5 = np.random.randint(100, size= (6,3))
print('Before transpose: \n',arr5)
print(' ')
print('After transpose: \n',np.transpose(arr5))
```

ðŸ“ƒ **5**. Create two random numpy array of size n x p and p x m and then compute the dot product


```python
arr6 = np.random.randint(10, size = (1,2))
print('Array 6 : \n', arr6)
arr7 = np.random.randint(10, size = (2,3))
print('Array 7 : \n',arr7)
print(' ')
print('Dot product: ', np.dot(arr6,arr7))
```
